### QuestGen.AI

#### Generate FAQs

```python
question=[]
answer=[]
context=[]
listing=[]
```

```python
def generateFAQ(payload_text,listing_id):
    payload = {
            "input_text": payload_text
    }
    output = qg.predict_shortq(payload)
    try:
        ques = output['questions']
        for q in ques:
            question.append(q['Question'])
            answer.append(q['Answer'])
            context.append(q['context'])
            listing.append(listing_id)
    except:
        print('Null Question')
```

```python
import time
st = time.time()
```

```python
reviews = reviews0[1000:3000]
# reviews = reviews0
```

```python
count_listing = 0
prev_listing = 0
for index in reviews.index:
    comments = reviews['comments'][index]
    listing_id = reviews['listing_id'][index]
    if(listing_id==prev_listing):
        count_listing+=1
    prev_listing = listing_id
    generateFAQ(str(comments),listing_id)
    
```

```python
data = { 
        'question':question,
        'answer':answer,
        'listing_id':listing,
        'context': context
}
df = pd.DataFrame(data)
df.to_csv(processor+'FAQ_2.json',index=None)
```
